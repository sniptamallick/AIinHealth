```python
from google.colab import files
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load patients data
d1 = files.upload()
patients = pd.read_csv('PATIENTS.csv')
```

```python
# Load procedures data
d2 = files.upload()
procedures_icd = pd.read_csv('PROCEDURES_ICD.csv')
```

```python
# Load procedures dictionary
d3 = files.upload()
procedures_d = pd.read_csv('D_ICD_PROCEDURES.csv')
```

```python
procedures_icd.head()
```

```python
# Merge procedures and patients data on subject_id
merged_df = pd.merge(patients, procedures_icd, on="subject_id")
merged_df.head(30)
```

```python
#seperate merged dataset into male and female
male_patients = merged_df[merged_df['gender'] == 'M']
female_patients= merged_df[merged_df['gender'] == 'F']
male_patients.head()
female_patients.head()
```

```python
# Count the frequency of procedures for female patients in the dataset
top_proceduresf = female_patients['icd9_code'].value_counts().head(10)
top_proceduresf
```

```python
# Count the frequency of procedures for male patients in the dataset
top_proceduresm = male_patients['icd9_code'].value_counts().head(10)
top_proceduresm
```

```python
#map the top procedures to the ICD9_procedures dictionary
proceduresf_labeled = pd.merge(top_proceduresf, procedures_d, on="icd9_code")
proceduresf_labeled
```

```python
#map the top procedures to the ICD9_procedures dictionary
proceduresm_labeled = pd.merge(top_proceduresm, procedures_d, on="icd9_code")
proceduresm_labeled
```

```python
#make two plots
f, axs = plt.subplots(1, 2, figsize = (15, 8))
a = sns.barplot(data = proceduresf_labeled, y = "short_title", x = "count", orient = "h", ax=axs[0], palette = "PuRd_r")
a.set(xlabel="Total", ylabel="ICD-9 Coded Procedure (Short Name)")
a.set_title("Top 10 Procedures for Female Patients")

b = sns.barplot(data = proceduresm_labeled, y = "short_title", x = "count", orient = "h", ax=axs[1], palette = "PuBu_r")
b.set(xlabel="Total", ylabel="ICD-9 Coded Procedure (Short Name)")
b.set_title("Top 10 Procedures for Male Patients")

plt.subplots_adjust(wspace=0.5)
```

```python
#visualize the lab values with a violin plot

#load lab events
d4 = files.upload()
labevents = pd.read_csv('LABEVENTS.csv')
```

```python
#load lab events
d5 = files.upload()
labevents_d = pd.read_csv('D_LABITEMS.csv')
```

```python
#merge datasets
labevents_patients = pd.merge(patients, labevents, on = "subject_id")
labevents_patients
```

```python
# join labevents_patients with labevents_d on the itemid column

labevents_patients_d = pd.merge(labevents_patients, labevents_d, on='itemid', how='left')
labevents_patients_d
```

```python
f, axs = plt.subplots(1, 2, figsize = (15, 8))
a = sns.violinplot(data=labevents_patients_d[(labevents_patients_d['label'] == 'White Blood Cells')],  x="label", y="valuenum", hue="gender", split=True,  gap=.1, inner="quart", ax=axs[0], palette = 'BuPu')
a.set(xlabel=" ", ylabel="White Blood Cell Count (K/uL)")
a.set_title("White Blood Cell Count Distribution")

b = sns.violinplot(data=labevents_patients_d[(labevents_patients_d['label'] == 'Red Blood Cells')],  x="label", y="valuenum", hue="gender", split=True,  gap=.1, inner="quart", ax=axs[1], palette = 'BuPu')
b.set(xlabel=" ", ylabel="Red Blood Cell Count (m/uL)")
b.set_title("Red Blood Cell Count Distribution")
```

```python
a = sns.kdeplot(data=labevents_patients_d[(labevents_patients_d['label'] == 'Hematocrit')], y="valuenum", hue = "gender", multiple = "stack")
plt.fill_between(x=[0.0, 0.1], y1=36, y2=44, color='blue', alpha=0.3)
plt.fill_between(x=[0.0, 0.1], y1=41, y2=55, color='orange', alpha=0.3)
a.set(xlabel="Density", ylabel="Hematocrit Percentage Value")
a.set_title("Hematocrit Distribution by Gender")
```

```python
#visualize the top 20 drugs using the prescriptions dataset
#upload the prescriptions dataset
d6 = files.upload()
prescriptions = pd.read_csv('PRESCRIPTIONS.csv')
```

```python
#select the top 20 precriptions
top_20_drugs = prescriptions['drug'].value_counts().head(20)
top_drugs = top_20_drugs.to_frame().reset_index()
top_drugs
```

```python
import plotly.express as px
```

```python
print(top_drugs.columns)
```

```python
px.treemap(
    top_drugs,
    path=['drug'],
    values='count',
    title='Top 20 Most Frequently Occurring Prescriptions',
    color = 'count',
    color_continuous_scale='Picnic'
)

```

```python
#create a sunburst plot to visualize mode of delivery for the top 20 prescriptions

#subset the main chart by the top 20 prescriptions
subset_df = prescriptions[prescriptions['drug'].isin(top_drugs['drug'])]
subset_df2 = pd.merge(subset_df, top_drugs, on = 'drug')
subset_df2.head(10)
```

```python
#drop duplicates
final_df = subset_df2.drop_duplicates(subset='drug')
final_df.head(10)
```

```python
#create sunburst plot

fig = px.sunburst(final_df, path=['route', 'drug'], values='count')
fig.show()
```
